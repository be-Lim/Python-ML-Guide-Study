```python
#임의의 데이터 생성
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)

X = 2 * np.random.rand(100,1)
y = 6 + 4 * X + np.random.randn(100,1)

plt.scatter(X, y)
```

```python
#비용 함수
def cost_func(y, y_pred):
    N = len(y)
    cost = np.sum(np.square(y - y_pred)) / N
    return cost
```

```python
#가중치 업데이트 함수
def update_weights(w0, w1, X, y, learning_rate):
    N = len(y)
    y_pred = w0 + w1 * X
    diff = y - y_pred

    w0_update = -learning_rate*(2/N)*(np.sum(diff))
    w1_update = -learning_rate*(2/N)*(np.dot(X.T, diff))

    return w0_update, w1_update
```

```python
#경사하강법 함수
def gradient_descent(X, y, iteration):
    N = len(y)
    #실수 주의: Parameter의 크기는 Feature 개수를 따라간다
    w0 = np.zeros((1,1))
    w1 = np.zeros((1,1))

    for i in range(iteration):
        w0_update, w1_update = update_weights(w0, w1, X, y, 0.01)
        w0 = w0 - w0_update
        w1 = w1 - w1_update

    return w0, w1
```

```python
#경사하강법 실행
w0, w1 = gradient_descent(X, y, 1000)
y_pred = w0 + w1 * X
cost = cost_func(y, y_pred)
print(f'w0: {w0[0,0]:.3f}, w1: {w1[0,0]:.3f}, cost: {cost:.3f}')
```
